```python
movie_info = defaultdict(dict)
def get_movie_info_fromlist(movie_list):
    for movie in movie_list:
        soup = BeautifulSoup(movie, 'xml')
        tv_movie_tag = str(soup(class_="infobar")[0].next)
        try:
            if re.sub(r'\W+', '', tv_movie_tag) != 'TVMovie' and re.sub(r'\W+', '', tv_movie_tag) != 'TVSpecial':
                ## get title
                head = soup(class_='header')
                sp = head[0](class_="itemprop")
                title = str(sp[0].string)
                print title
            else:
                title = 'TVMovie'
        except UnicodeEncodeError:
            title = 'Error'
            print title
        ## get date
        try:
            if soup('h4')[20].contents == [u'Release Date:']:
                date_string = str(soup('h4')[20].next.next)
            elif soup('h4')[19].contents == [u'Release Date:']:
                date_string = str(soup('h4')[19].next.next)
            elif soup('h4')[17].contents == [u'Release Date:']:
                date_string = str(soup('h4')[17].next.next)    
            else:
                date_string = str(soup('h4')[18].next.next)
            date_split = date_string.split( ' ')
            date = date_split[1] + ' ' + date_split[2] + ' ' + date_split[3]
        except:
            date = 'Error'
        movie_info[title]['date']=date
        ##get keywords
        try: 
            url = str(soup(property="og:url")[0]).split ('"')[1]
            key_word_url_ending = 'keywords?ref_=tt_stry_kw'
            key_words_url = url + key_word_url_ending
            key_response = requests.get(key_words_url)
            key_page = key_response.text
            key_soup = BeautifulSoup(key_page, 'xml')
            keyword_list = []
            for item in key_soup.find_all(class_="soda sodavote"):
                keyword_list.append(str(item.contents[1].contents[1].contents[0]))
        except:
            keyword_list = 'Error'
        movie_info[title]['keywords'] = keyword_list
        ##get genres 
        genre_tags = soup('span', itemprop="genre")
        genre_list = []
        for item in genre_tags:
            genre = str(item.string)
            genre_list.append(genre)
        if title in sequels:
            genre_list.append('sequel')
        movie_info[title]['genre']=genre_list
        try:
            budget_string = soup(text="Budget:")[0].next
            budget = int(re.sub("[^0-9]", "", budget_string))
        except: 
            budget = 'Error'
        movie_info[title]['budget'] = budget
        try:
            box_office_href = 'business?ref_=tt_dt_bus'
            url = str(soup(property="og:url")[0]).split ('"')[1]
            box_office_url = url + box_office_href
            ## go to box office info
            box_response = requests.get(box_office_url)
            box_page = box_response.text
            box_soup = BeautifulSoup(box_page, 'xml')
            ## get opening weekend amount
            if box_soup(text='Opening Weekend'):
                if '$' in box_soup(text='Opening Weekend')[0].next:
                    opening_string = str(box_soup(text='Opening Weekend')[0].next)
                    opening_weekend = int(re.sub("[^0-9]", "", opening_string))
                else: opening_weekend = 'N/A'
            else: opening_weekend = 'N/A'
        except UnicodeEncodeError:
            opening_weekend = 'Error'
        movie_info[title]['opening']=opening_weekend
        try:
            screen_string = box_soup(text='Opening Weekend')[0].next.next.next.next.next.next.next           
            screens = str(re.sub(r'[^1-9]', '', screen_string))
        except:
            screens = 'Error'
        movie_info[title]['screens'] = screens
```
