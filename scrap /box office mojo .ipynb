```python
def date_helper(date1, date2):
    try:
        date1 = parser.parse(date1)
        date2 = parser.parse(date2, fuzzy=True)
        delta = abs(date1 - date2)
        delta_int = delta.days
        if delta_int < 7:
            return True
    except: 
        pass
    return False
```
