```python
title_list = soup.find_all(class_='title')
```

```python
href_list = []
count = 0
for title in title_list:
    while count < 250:
        a = title_list[count].contents
        href = a[0].get('href')
        imdb = 'http://www.imdb.com'
        movie_url = str(imdb + href)
        href_list.append(movie_url)
        count += 1
        
```
